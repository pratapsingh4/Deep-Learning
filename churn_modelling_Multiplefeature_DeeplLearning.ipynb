```python
import pandas a
s pd
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
y=df.iloc[:,-1]
```

```python
x=df.iloc[:,3:-1]
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
labelencoder_X_geo =LabelEncoder()
```

```python
x.iloc[:,1]=labelencoder_X_geo.fit_transform(x.iloc[:,1])
```

```python
labelencoder_X_gen =LabelEncoder()
```

```python
x.iloc[:,2]=labelencoder_X_gen.fit_transform(x.iloc[:,2])
```

```python
x
```

```python
onehotencoder_geo=OneHotEncoder(categorical_features=[1]) #position 1
```

```python
X=onehotencoder_geo.fit_transform(x).toarray() # to convert numpy as sklearn uses numpy
```

```python
X=X[:,1:]   #here X is numpy array
            # use iloc in dataframe
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from keras .models import Sequential
from keras.layers import Dense
```

```python
classifier=Sequential()
```

```python
X.shape   # has 11 features
```

```python
# single hidden layer #
classifier.add(Dense(units=6,input_shape=(11,),activation='relu'))
#11+1/2=6 generally
#relu when output is more like 5 or 6
# second hidden layer #
#outut or first and second layer is given genrally input+output/2(11+1/2)
classifier.add(Dense(units=6,activation='relu'))
# third hidden layer #
classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
from keras.optimizers import Adam,SGD
```

```python
classifier.compile(Adam(lr=0.5),loss="binary_crossentropy",metrics=['accuracy'])
```

```python
classifier.fit(X_train,y_train,epochs=10)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred=(y_pred >0.5)
```

```python
y_pred     #27 percent for not leaving
```

```python
 from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```
